```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)                     # Scatter means print x verses y term
```

```python
colours  = np.random.rand(50)         #print different colours
sizes = 1000* np.random.rand(50)
plt.scatter(x,y, c = colours , s = sizes , alpha =.1)
plt.xlabel("x axis")
plt.ylabel("y axix")
plt.title("random data")
```
